```python
df_Confirmed = pd.read_csv('time_series_19-covid-Confirmed.csv')
df_Deaths = pd.read_csv('time_series_19-covid-Deaths.csv')
df_Recovered = pd.read_csv('time_series_19-covid-Recovered.csv')
```

# Confirmed Cases

```python
country = 'US'
temp_df = df_Confirmed[df_Confirmed['Country/Region'] == country].copy()
temp_df.drop(['Province/State'], axis = 1, inplace= True)
temp_df.drop(['Lat'], axis = 1, inplace= True)
temp_df.drop(['Long'], axis = 1, inplace= True)
temp_df.drop(['Country/Region'], axis = 1, inplace= True)
temp_df = temp_df.T.copy()
temp_df['US Total Confirmed'] = 0

for index in temp_df.index:
    temp_df.loc[index, 'US Total Confirmed'] = sum(temp_df.loc[index,])
temp_df = temp_df[['US Total Confirmed']].copy()
same_day_table = temp_df.copy()

for country in ['Japan', 'Italy','China','Iran','Spain','Korea, South']:

    temp_df = df_Confirmed[df_Confirmed['Country/Region'] == country].copy()
    temp_df.drop(['Province/State'], axis = 1, inplace= True)
    temp_df.drop(['Lat'], axis = 1, inplace= True)
    temp_df.drop(['Long'], axis = 1, inplace= True)
    temp_df.drop(['Country/Region'], axis = 1, inplace= True)
    temp_df = temp_df.T.copy()
    temp_name = country + ' Total Confirmed'
    temp_df[temp_name] = 0

    for index in temp_df.index:
        temp_df.loc[index, temp_name] = sum(temp_df.loc[index,])
    temp_df = temp_df[[temp_name]].copy()

    same_day_table = pd.merge(same_day_table,temp_df,left_index=True , right_index= True, how = 'outer')

same_day_table
```

# Deaths Cases

```python
for country in ['US','Japan', 'Italy','China','Iran','Spain','Korea, South']:

    temp_df = df_Deaths[df_Deaths['Country/Region'] == country].copy()
    temp_df.drop(['Province/State'], axis = 1, inplace= True)
    temp_df.drop(['Lat'], axis = 1, inplace= True)
    temp_df.drop(['Long'], axis = 1, inplace= True)
    temp_df.drop(['Country/Region'], axis = 1, inplace= True)
    temp_df = temp_df.T.copy()
    temp_name = country + ' Total Deaths'
    temp_df[temp_name] = 0

    for index in temp_df.index:
        temp_df.loc[index, temp_name] = sum(temp_df.loc[index,])
    temp_df = temp_df[[temp_name]].copy()

    same_day_table = pd.merge(same_day_table,temp_df,left_index=True , right_index= True, how = 'outer')

same_day_table
```

# Recovered Cases

```python
for country in ['US','Japan', 'Italy','China','Iran','Spain','Korea, South']:

    temp_df = df_Recovered[df_Recovered['Country/Region'] == country].copy()
    temp_df.drop(['Province/State'], axis = 1, inplace= True)
    temp_df.drop(['Lat'], axis = 1, inplace= True)
    temp_df.drop(['Long'], axis = 1, inplace= True)
    temp_df.drop(['Country/Region'], axis = 1, inplace= True)
    temp_df = temp_df.T.copy()
    temp_name = country + ' Total Recovered'
    temp_df[temp_name] = 0

    for index in temp_df.index:
        temp_df.loc[index, temp_name] = sum(temp_df.loc[index,])
    temp_df = temp_df[[temp_name]].copy()

    same_day_table = pd.merge(same_day_table,temp_df,left_index=True , right_index= True, how = 'outer')

same_day_table
```

# Mortality/Recovery Rate

```python
for country in ['US','Japan', 'Italy','China','Iran','Spain','Korea, South']:
    temp_name_Mortality = country + ' Mortality Rate'
    temp_name_Recovery = country + ' Recovery Rate'
    
    temp_name_Confirmed = country + ' Total Confirmed'
    temp_name_Deaths = country + ' Total Deaths'
    temp_name_Recovered = country + ' Total Recovered'
    
    same_day_table[temp_name_Mortality] = same_day_table[temp_name_Deaths] / same_day_table[temp_name_Confirmed]
    same_day_table[temp_name_Recovery] = same_day_table[temp_name_Recovered] / same_day_table[temp_name_Confirmed]

same_day_table
```

```python
same_day_table.to_csv('time_series_19-covid-same_day_overall.csv')
```
